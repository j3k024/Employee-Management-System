```python
list_of_employees = []

# --- 1. Function Definitions ---

def add_employee():
    check = True
    print('\n--- Enter Employee Details ---')
    id = int(input('ID: '))
    name = input('Name: ')
    age = input('Age: ')
    position = input('Position: ')
    salary = int(input('Salary: '))

    # Check for duplicate ID
    if not list_of_employees:
        check = True
    else:
        for employee in list_of_employees:
            if id == employee['id']:
                check = False
                break

    if check:
        employee = {
            'id': id,
            'name': name,
            'age': age,
            'position': position,
            'salary': salary
        }
        list_of_employees.append(employee)
        print('Employee added successfully!')
    else:
        print('Error: ID already exists.')

def update_employee():
    if not list_of_employees:
        print("No employees to update.")
        return

    id = int(input('Enter ID to update: '))
    for employee in list_of_employees:
        if id == employee['id']:
            print('\nWhich information you want to update:')
            print('1. Name')
            print('2. Position')
            print('3. Salary')
            ch = input('Select choice (1-3): ')

            if ch == '1':
                print(f"Current Name: {employee['name']}")
                employee['name'] = input('Enter new name: ')
            elif ch == '2':
                print(f"Current Position: {employee['position']}")
                employee['position'] = input('Enter new position: ')
            elif ch == '3':
                print(f"Current Salary: {employee['salary']}")
                # Convert to int to keep data type consistent
                employee['salary'] = int(input('Enter new salary: '))
            else:
                print('Invalid Choice.')
                return

            print('Employee details updated successfully!')
            return
    print('Employee not found.')

def delete_employee():
    if not list_of_employees:
        print("No employees to delete.")
        return

    id = int(input('Enter ID to delete: '))
    for employee in list_of_employees:
        if id == employee['id']:
            list_of_employees.remove(employee)
            print('Employee deleted successfully!')
            return
    print('Employee not found.')

def display_employee():
    if len(list_of_employees) == 0:
        print('\nNo employees to display.')
        return

    print('\n' + '-'*70)
    # Using padding (<10) to make columns align perfectly
    print(f"{'ID':<10} {'Name':<15} {'Age':<10} {'Position':<15} {'Salary':<10}")
    print('-'*70)
    for emp in list_of_employees:
        print(f"{emp['id']:<10} {emp['name']:<15} {emp['age']:<10} {emp['position']:<15} {emp['salary']:<10}")
    print('-'*70)

# --- 2. Main Menu Loop ---

while True:
    print('\n==============================')
    print('  Employee Management System')
    print('==============================')
    print('1. Add Employee')
    print('2. Update Employee')
    print('3. Delete Employee')
    print('4. Display Employee')
    print('5. Exit')
    print('------------------------------')

    choice = input('Select an option: ')

    if choice == '1':
        add_employee()
    elif choice == '2':
        update_employee()
    elif choice == '3':
        delete_employee()
    elif choice == '4':
        display_employee()
    elif choice == '5':
        print('Exiting program. Goodbye!')
        break
    else:
        print('Invalid option, please try again.')
```
